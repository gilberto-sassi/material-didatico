```r
if (!require(pacman)) {
    install.packages('pacman')
    library(pacman)
}
p_load(
    readxl, writexl, janitor, googlesheets4, tidyverse
)
```

```r
positivo <- seq(from = 0, to = 4, by = 0.01)
```

```r
df <- tibble(positivo)
glimpse(df)
```

```r
df <- df |>
  mutate(prob = map_dbl(positivo, ~ pnorm(.x)) |> round(4)) |>
  mutate(
    inteiro = floor(positivo),
    decimal = positivo - floor(positivo)
  )
```

```r
glimpse(df)
```

```r
write_sheet(
  df,
  "https://docs.google.com/spreadsheets/d/1xi5fbjrRrnIqokv6UKEfAZNP7kSPeWv0YkC4U_zlfkY/edit?usp=sharing",
  sheet = "positivo"
)
```

```r
negativo <- seq(from = -4, to = 0, by = 0.01)
```

```r
df <- tibble(negativo) |>
  mutate(prob = map_dbl(negativo, ~ pnorm(.x) |> round(4))) |>
  mutate(
    inteiro = ceiling(negativo),
    decimal =  (negativo - ceiling(negativo))
  )
```

```r
glimpse(df)
```

```r
write_sheet(
  df,
  "https://docs.google.com/spreadsheets/d/1xi5fbjrRrnIqokv6UKEfAZNP7kSPeWv0YkC4U_zlfkY/edit?usp=sharing",
  sheet = "negativo"
)
```
