```r
if (!require(pacman)) {
  install.packages("pacman")
  library(pacman)
}
p_load(
    readxl, writexl, tidyverse
)
```

```r
dados <- read_excel("dados_atv.xlsx")
dados
```

```r
suporte <- dados$suporte^2 |> unique() |> sort()
suporte
```

```r
n <- length(suporte) - 1
fp <- vector("numeric", length(suporte))
fp[1] <- dados$fp[n + 1]
for (k in 1:n) {
  fp[k + 1] = dados$fp[n + 1 - k] + dados$fp[n + 1 + k]
}
fp
```

```r
sum(fp)
```

```r
resposta <- tibble(
  suporte, fp
)
```

```r
resposta
```

```r
write_xlsx(resposta, "resposta.xlsx")
```
