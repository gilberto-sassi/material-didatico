```python
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
tamanho = 16
media = 5
df_poisson = pd.DataFrame({
    "x": list(range(tamanho + 1)),
    "fp": np.round( st.poisson.pmf(range(tamanho + 1), media)  , 4),
    "fda": np.round( st.poisson.cdf(range(tamanho + 1), media)  , 4)
})
df_poisson
```

```python
# binomial negativa
p = 0.55
tamanho = 13
n = 1
df_bn = pd.DataFrame({
    "x": list(range(tamanho)),
    "fp": np.round(st.nbinom.pmf(range(tamanho), n, p), 4),
    "fda": np.round(st.nbinom.cdf(range(tamanho), n, p), 4)
})
df_bn
```
