```r
install.packages("pacman")
library(pacman)
p_load(
    readxl, writexl, janitor, tidyverse
)
```

```r
f <- \(x) format(x, decimal.mark = ",", big.mark = ".")
# binomial
pbinom(20, 76, 0.29) |> f()
(pbinom(26, 76, 0.29) - pbinom(19, 76, 0.29)) |> f()
```

```r
# normal
f(pnorm(12.65, 11.32, 2.1))
f(pnorm(13.24, 11.32, 2.1) - pnorm(8.936, 11.32, 2.1))
```

```r
# poisson
f(ppois(13, 12))
f(ppois(15, 12) - ppois(9, 12))
```

```r
# gamma
f(pgamma(0.3068, 1, 1))
f(pgamma(1.736, 1, 1) - pgamma(0.2191, 1, 1))
```

```r
# exponencial
f(pexp(1.016, 1))
f(pexp(1.583, 1) - pexp(0.1936022, 1))
```

```r
# uniforme
f(punif(13.77, 8.388, 19.37))
f(punif(17.11, 8.388, 19.37) - punif(10.13, 8.388, 19.37))
```

```r
# beta
f(pbeta(0.1229, 1.085, 8.331))
f(pbeta(0.177, 1.085, 8.331) - pbeta(0.02655, 1.085, 8.331))
```

```r
# lognormal
f(plnorm(2.779, 1.178, 0.9136))
f(plnorm(6.288, 1.178, 0.9136) - plnorm(1.301, 1.178, 0.9136))
```
