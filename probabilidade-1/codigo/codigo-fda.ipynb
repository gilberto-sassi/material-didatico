```r
f <- \(x) format(x, decimal.mark = ",", big.mark = ".")
# binomial
n <- 62
p <- 0.3
x <- 16
b <- 22
a <- 15
pbinom(x, n, p) |> f()
(pbinom(b, n, p) - pbinom(a, n, p)) |> f()
```

```r
# normal
media <- 14.23
s <- 2.61
x <- 12.53
b <- 16.85
a <- 11.94
f(pnorm(x, media, s))
f(pnorm(b, media, s) - pnorm(a, media, s))
```

```r
# poisson
taxa <- 14.48
x <- 14
b <- 19
a <- 11
f(ppois(x, taxa))
f(ppois(b, taxa) - ppois(a, taxa))
```

```r
# gamma
pa <- 1
pb <- 1
x <- 1.148
b <- 1.617
a <- 0.1261
f(pgamma(x, pa, pb))
f(pgamma(b, pa, pb) - pgamma(a, pa, pb))
```

```r
# exponencial
taxa <- 1
x <- 0.3052
b <- 1.217
a <- 0.1766
f(pexp(x, taxa))
f(pexp(b, taxa) - pexp(a, taxa))
```

```r
# uniforme
maximo <- 18.76
minimo <- 7.277
x <- 11.49
b <- 16.44
a <- 9.062
f(punif(x, minimo, maximo))
f(punif(b, minimo, maximo) - punif(a, minimo, maximo))
```

```r
# beta
shape1 <- 21.19
shape2 <- 7.893
x <- 0.7745
b <- 0.7889
a <- 0.6373
f(pbeta(x, shape1, shape2))
f(pbeta(b, shape1, shape2) - pbeta(a, shape1, shape2))
```

```r
# lognormal
pa <- 0.3332
pb <- 0.07824
x <- 1.335
b <- 1.47
a <- 1.303
f(plnorm(x, pa, pb))
f(plnorm(b, pa, pb) - plnorm(a, pa, pb))
```
