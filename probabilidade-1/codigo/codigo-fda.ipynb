```r
f <- \(x) format(x, decimal.mark = ",", big.mark = ".")
# binomial
n <- 83
p <- 0.51
x <- 42
b <- 47
a <- 37
pbinom(x, n, p) |> f()
(pbinom(b, n, p) - pbinom(a, n, p)) |> f()
```

```r
# normal
media <- 11.29
s <- 2.06
x <- 12.58
b <- 13.23
a <- 9.18
f(pnorm(x, media, s))
f(pnorm(b, media, s) - pnorm(a, media, s))
```

```r
# poisson
taxa <- 14.92
x <- 14
b <- 20
a <- 0
f(ppois(x, taxa))
f(ppois(b, taxa) - ppois(a, taxa))
```

```r
# gamma
pa <- 1
pb <- 1
x <- 1.375
b <- 2.098
a <- 0.2034
f(pgamma(x, pa, pb))
f(pgamma(b, pa, pb) - pgamma(a, pa, pb))
```

```r
# exponencial
taxa <- 1
x <- 1.035
b <- 1.404
a <- 0.1788
f(pexp(x, taxa))
f(pexp(b, taxa) - pexp(a, taxa))
```

```r
# uniforme
minimo <- 5.766
maximo <- 18.71
x <- 9.773
b <- 15.73
a <- 7.778
f(punif(x, minimo, maximo))
f(punif(b, minimo, maximo) - punif(a, minimo, maximo))
```

```r
# beta
shape1 <- 3.278
shape2 <- 0.5602
x <- 0.09905
b <- 0.197
a <- 0.008205
f(pbeta(x, shape1, shape2))
f(pbeta(b, shape1, shape2) - pbeta(a, shape1, shape2))
```

```r
# lognormal
pa <- 1.419
pb <- 0.05483
x <- 4.043
b <- 4.277
a <- 3.914
f(plnorm(x, pa, pb))
f(plnorm(b, pa, pb) - plnorm(a, pa, pb))
```
