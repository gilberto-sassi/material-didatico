# nossas próprias funções

````python
def tabela(data_frame, nome_coluna):
  frequencia = data_frame.loc[:, nome_coluna].value_counts()
  frequencia_relativa = data_frame.loc[:, nome_coluna].value_counts(normalize=True)
  porcentagem = frequencia_relativa * 100
  tab = pd.DataFrame({
    'frequencia': frequencia,
    'frequencia_relativa': frequencia_relativa,
    'porcentagem': porcentagem
  })
  return tab
````

````python
tab_distribuicao = tabela(enem, 'tp_sexo')
````

````python
tab_distribuicao
````

# gráficos

## gráfico de barras

Variável qualitativa

* cada categoria tem um barra
* existe um espaço entre as barras
* a altura da barra é frequência (frequência relativa, porcentagem)



````python
# padrão é usar a frequência absoluta (contagem)
fig, ax = plt.subplots()
sns.countplot(data=enem, x='tp_sexo', ax=ax, color='#6699ff') # color pode ser cor html
ax.set_ylabel('Frequência')
ax.set_xlabel('Gênero')
plt.show()
````

````python
# padrão é usar a frequência absoluta (contagem)
fig, ax = plt.subplots()
sns.countplot(data=enem, x='tp_sexo', ax=ax, color='#6699ff', stat='percent') # color pode ser cor html
ax.set_ylabel('Porcentagem')
ax.set_xlabel('Gênero')
plt.show()
fig.savefig('figuras/genero.png')
fig.savefig('figuras/genero.jpeg')
fig.savefig('figuras/genero.pdf')
````
