```python
def tabela(dados, variavel):
  return pd.DataFrame({
      'frequencia': dados[[variavel]].value_counts(),
      'freq_rel': dados[[variavel]].value_counts(normalize=True),
      'porcentagem': dados[[variavel]].value_counts(normalize=True) * 100
  })
```

```python
tab_cor_raca = tabela(enem, 'tp_cor_raca')
```

```python
tab_cor_raca.to_excel('tabelas/tab_tp_cor_raca.xlsx')
```

## gráficos

```python
tab = enem.tp_cor_raca.value_counts(ascending=False, normalize=True)
tab.index
```

```python
fig, ax = plt.subplots()
sns.countplot(data=enem, y='tp_cor_raca', ax=ax, stat='proportion', order=tab.index)
ax.set_xlabel('Frequência')
ax.set_ylabel('Cor/Raça')
ax.set_xlim([0, 1])
ax.bar_label(container=ax.containers[0], labels=np.round(tab * 100, 2))
plt.show()
fig.savefig('figuras/tp_cor_raca.png')
```

```python
tab = enem['q005'].value_counts(normalize=True).sort_index()
fig, ax = plt.subplots(figsize=(12, 10))
sns.countplot(data=enem, y='q005', ax=ax, stat='count', order=tab.index)
ax.set_xlabel('Frequência')
ax.set_ylabel('Q005')
#ax.set_xlim([0, 1])
ax.bar_label(container=ax.containers[0],
             labels=[f'{v * 100:.2f} %'.replace(".", ",") for v in tab])
plt.show()
fig.savefig('figuras/q005.png')
```

### Histograma

```python
k = np.ceil(1 + np.log2(enem.shape[0])).astype(np.int64)
fig, ax = plt.subplots()
sns.histplot(data=enem, x='nu_nota_mt',
             bins=k,
             stat='density', ax=ax)
ax.set_xlabel('Nota de matemática')
ax.set_ylabel('Densidade de frequência')
plt.show()
fig.savefig('figuras/hist_nu_nota_mt.png')
```

```python
fig, ax = plt.subplots()
sns.histplot(data=enem, x='nu_nota_mt',
             binwidth=50,
             stat='density', ax=ax)
ax.set_xlabel('Nota de matemática')
ax.set_ylabel('Densidade de frequência')
plt.show()
```

```python
fig, ax = plt.subplots()
sns.histplot(data=enem, x='nu_nota_lc',
             binwidth=50,
             stat='count', ax=ax)
ax.set_xlabel('Nota de matemática')
ax.set_ylabel('Densidade de frequência')
plt.show()
```

```python
fig, ax = plt.subplots()
sns.histplot(data=enem, x='nu_nota_lc',
             binwidth=50,
             stat='percent', ax=ax)
ax.set_xlabel('Nota de matemática')
ax.set_ylabel('Densidade de frequência')
plt.show()
```
